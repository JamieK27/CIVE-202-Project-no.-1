```python
#Task 1 Code:  
#Extract and organize AirPurple Data

#import pandas package
import pandas as pd

#load data
air_data = pd.read_csv('AirQuality_Daily_StudentVersion.csv')

#turn data into a dataframe
air_data = pd.DataFrame(air_data)

#checking data
#air_data.head(5)

```

```python
#Task 2 Code: 
#Determine 5 locations in Nebraska with the highest mean and median concentrations of VOC, PM 2.5, Pm 10.0
#for pm2.5, pm10.0, and voc

#voc mean and median
stat_summary_voc = air_data.groupby(['sensor.name']).agg({'voc': ['mean', 'median']})

top_5_voc_mean = (stat_summary_voc.sort_values( by = ('voc', 'mean'), ascending = False)).head(5)
top_5_voc_mean

top_5_voc_median = (stat_summary_voc.sort_values( by = ('voc', 'median'), ascending = False)).head(5)
top_5_voc_median 
```

```python
#Task 2 Continued:
#pm 2.5 mean and median
stat_summary_pm25 = air_data.groupby(['sensor.name']).agg({'pm2.5_atm': ['mean', 'median']})

top_5_pm25_mean = (stat_summary_pm25.sort_values( by = ('pm2.5_atm', 'mean'), ascending = False)).head(5)
top_5_pm25_mean

top_5_pm25_median = (stat_summary_pm25.sort_values( by = ('pm2.5_atm', 'median'), ascending = False)).head(5)
top_5_pm25_median 
```

```python
#Task 2 Continued:
#pm 10.0 mean and median
stat_summary_pm10 = air_data.groupby(['sensor.name']).agg({'pm10.0_atm': ['mean', 'median']})

top_5_pm10_mean = (stat_summary_pm10.sort_values( by = ('pm10.0_atm', 'mean'), ascending = False)).head(5)
top_5_pm10_mean

top_5_pm10_median = (stat_summary_pm10.sort_values( by = ('pm10.0_atm', 'median'), ascending = False)).head(5)
top_5_pm10_median 
```

```python
#Task 3 Code:  
#Determine what days maximum values occurred and the location
#voc, date, location
max_voc = air_data.sort_values('voc', ascending=False).head(1)
max_voc[['date', 'sensor.name', 'voc']]
```

```python
#Task 3 Continued:
#pm 2.5, date, location
max_pm25 = air_data.sort_values('pm2.5_atm', ascending=False).head(1)
max_pm25[['date', 'sensor.name', 'pm2.5_atm']]
```

```python
#Task 3 Continued:
#pm 10.0, date, location
max_pm10 = air_data.sort_values('pm10.0_atm', ascending=False).head(1)
max_pm10[['date', 'sensor.name', 'pm10.0_atm']]
```

```python
#Task 4 Code:
#Determine the relation of air quality, humidity, and temperature

#create humidity categories
def assign_humidity(x):
    if x < 50:
        return 'Less than 50%'
    elif 50 <= x <= 80:
        return 'Between 50% and 80%'
    else:
        return 'Above 80%'

air_data['humidity_level'] = air_data['humidity'].apply(assign_humidity)
air_data[['sensor.name', 'humidity_level', 'voc', 'pm2.5_atm', 'pm10.0_atm']].head(15)

#high humidity
high_humidity = air_data[(air_data['humidity_level'] == 'Above 80%')].groupby('humidity')
high_voc_median = high_humidity.agg(median=('voc', 'median'))
high_vocs = high_voc_median.sort_values(by = 'median', ascending = False)
high_vocs.head(5)
```

```python
#Task 4 Code Continued:
#Middle humidity and voc
medium_humidity = air_data[(air_data['humidity_level'] == 'Between 50% and 80%')].groupby('humidity')
medium_voc_median = medium_humidity.agg(median=('voc', 'median'))
medium_vocs = medium_voc_median.sort_values(by = 'median', ascending = False)
medium_vocs.head(5)
```

```python
#Task 4 Code Continued:
#low humidity and voc
low_humidity = air_data[(air_data['humidity_level'] == 'Less than 50%')].groupby('humidity')
low_voc_median = low_humidity.agg(median=('voc', 'median'))
low_vocs = low_voc_median.sort_values(by = 'median', ascending = False)
low_vocs.head(5)
```

```python
#Task 4 Continued:

#create temperature catagories
def assign_temperature(x):
    if x < 32:
        return 'Below Freezing'
    elif 32 <= x <= 50:
        return 'Cool'
    elif 51 <= x <= 70:
        return 'Warm'
    else:
        return 'Hot'

#adding the column and checking that it worked
air_data['temperature_level'] = air_data['temperature'].apply(assign_temperature)
air_data[['sensor.name', 'temperature_level', 'temperature', 'voc', 'pm2.5_atm', 'pm10.0_atm']].head(15)

#below freezing
below_freezing = air_data[(air_data['temperature_level'] == 'Below Freezing')].groupby('temperature')
below_voc_median = below_freezing.agg(median=('voc', 'median'))
below_freezing_vocs = below_voc_median.sort_values(by = 'median', ascending = False)
below_freezing_vocs.head(5)
```

```python
#Task 4 Continued:

#Cool
cool_data = air_data[(air_data['temperature_level'] == 'Cool')].groupby('temperature')
cool_voc_median = cool_data.agg(median=('voc', 'median'))
cool_vocs = cool_voc_median.sort_values(by = 'median', ascending = False)
cool_vocs.head(5)
```

```python
#Task 4 Continued:

#Warm
warm_data = air_data[(air_data['temperature_level'] == 'Warm')].groupby('temperature')
warm_voc_median = warm_data.agg(median=('voc', 'median'))
warm_vocs = warm_voc_median.sort_values(by = 'median', ascending = False)
warm_vocs.head(5)
```

```python
#Task 4 Continued:

#Hot
hot_data = air_data[(air_data['temperature_level'] == 'Hot')].groupby('temperature')
hot_voc_median = hot_data.agg(median=('voc', 'median'))
hot_vocs = hot_voc_median.sort_values(by = 'median', ascending = False)
hot_vocs.head(5)
```

```python
#Task 5 Code: 
#Review data for any Air Quality Index health risks for PM 2.5 and PM 10.0

#PM 2.5
#Unhealthy for sensitive groups is 35.5-55.4 for PM 2.5 and 155-254 for PM 10.0


#defining a function to make unhealthy for sensitive groups a category
def assign_AQI(x):
    if x < 35.4:
        return "Moderate to Good Quality"
    elif 35.5 < x < 55.4:
        return "Unhealthy for Sensitive Groups"
    else:
        return "Unhealthy to Extremely Unhealthy"
    
#adding the column and checking that it worked again
air_data['pm2.5_level'] = air_data['pm2.5_atm'].apply(assign_AQI)
air_data[['sensor.name', 'pm2.5_level', 'pm2.5_atm']].head(15)

#PM 2.5 Unhealthy for Sensitive Groups Events
unhealthy_pm25_data = air_data[(air_data['pm2.5_level'] == 'Unhealthy for Sensitive Groups')].groupby('date')
unhealthy_pm25_values = unhealthy_pm25_data.agg(pm25 =('pm2.5_atm', 'median'))
unhealthy_pm25_values
```

```python
#Task 5 Code Continued:
#defining a function for PM 10.0
def assign_AQI_pm10(x):
    if x < 154:
        return "Moderate to Good Quality"
    elif 155 < x < 254:
        return "Unhealthy for Sensitive Groups"
    else:
        return "Unhealthy to Extremely Unhealthy"
    
#adding the column and checking that it worked again
air_data['pm10.0_level'] = air_data['pm10.0_atm'].apply(assign_AQI_pm10)
air_data[['sensor.name', 'pm10.0_level', 'pm10.0_atm']]

#PM 10.0 Unhealthy for Sensitive Groups Events
unhealthy_pm10_data = air_data[(air_data['pm10.0_level'] == 'Unhealthy for Sensitive Groups')].groupby('date')
unhealthy_pm10_values = unhealthy_pm10_data.agg(pm10=('pm10.0_atm', 'median'))
unhealthy_pm10_values
```
